```python
import requests

# URL de la API para obtener los assets del release
REPO = "DataTalksClub/nyc-tlc-data"
TAG = "fhv"
URL = f"https://api.github.com/repos/{REPO}/releases/tags/{TAG}"

# Hacer la petici√≥n a la API de GitHub
response = requests.get(URL)
if response.status_code == 200:
    release_data = response.json()
    
    # Extraer los archivos del release
    assets = release_data.get("assets", [])
    for asset in assets:
        print(f"üìÇ Archivo: {asset['name']}")
        print(f"üîó URL de descarga: {asset['browser_download_url']}\n")
else:
    print(f"‚ùå Error: No se pudo obtener la lista de archivos. C√≥digo {response.status_code}")
```

```python
import os

# Seleccionamos un archivo a descargar
TEST_FILE_URL = "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/fhv/fhv_tripdata_2019-01.csv.gz"
FILE_NAME = os.path.basename(TEST_FILE_URL)

# Descargar el archivo
response = requests.get(TEST_FILE_URL, stream=True)
if response.status_code == 200:
    with open(FILE_NAME, "wb") as f:
        for chunk in response.iter_content(chunk_size=8192):
            f.write(chunk)
    print(f"‚úÖ Archivo descargado: {FILE_NAME}")
else:
    print(f"‚ùå Error al descargar el archivo. C√≥digo {response.status_code}")
```

```python
from google.cloud import storage
import os
from tqdm import tqdm

# Configuraci√≥n de las credenciales y el bucket
GCS_CREDENTIALS_PATH = "/home/maxkaizo/DE_Zoomcamp/module_1/terraform_gcp/terrademo/keys/my-creds.json"
GCS_BUCKET_NAME = "dataeng-448500-hw-4"  # Aseg√∫rate de cambiarlo al nombre correcto

# Archivo de prueba
TEST_FILE_PATH = "image-1.png"
FILE_NAME_IN_GCS = "test_upload/image-1.png"

# Crear cliente con autenticaci√≥n expl√≠cita
client = storage.Client.from_service_account_json(GCS_CREDENTIALS_PATH)

def upload_to_gcs(file_path, file_name):
    """Funci√≥n para subir un archivo a GCS con barra de progreso."""
    bucket = client.bucket(GCS_BUCKET_NAME)
    blob = bucket.blob(file_name)

    total_size = os.path.getsize(file_path)
    with open(file_path, "rb") as f, tqdm(
        desc=f"üöÄ Subiendo {file_name}",
        total=total_size,
        unit="B",
        unit_scale=True,
        unit_divisor=1024,
        leave=True
    ) as bar:
        blob.upload_from_file(f, size=total_size, client=client)
        bar.update(total_size)

    print(f"‚úÖ Archivo subido: gs://{GCS_BUCKET_NAME}/{file_name}")

# Ejecutar la prueba de carga
upload_to_gcs(TEST_FILE_PATH, FILE_NAME_IN_GCS)
```
